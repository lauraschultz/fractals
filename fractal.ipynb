```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
origin_pts = np.array([ [0,0], [1, 0], [0.5, 1]])
```

```python
def gen_next(ar):
    n = np.random.randint(0,3)
    #     ori = origin_pts[n][0]
    #     half_x = (origin_pts[n][0] - x)/2
    #     half_y = (origin_pts[n][1] - y)/2
#     print(np.array(origin_pts[n]))
#     print(np.array(ar))
#     print([np.array(origin_pts[n]) - np.array(ar)])
    ar=np.array(ar)
    return  ar + ((np.array(origin_pts[n]) - ar)/2)
```

```python
plt.clf()
plt.figure(figsize=(15,15))
plt.ylim(-.1, 1.1)
plt.xlim(-.1, 1.1)
plt.xticks([])
plt.yticks([])
#base case
curr_pt = [.5,.5]
plt.scatter([curr_pt[0]], [curr_pt[1]], c='darkorange', s=6)
plt.scatter(origin_pts[...,0], origin_pts[...,1], color='steelblue', s=22)
curr_pt = gen_next(curr_pt)
plt.title('i = 1', fontsize=50)
plt.savefig('frames/00.png')
for p in range(0,5):
    for counter in range(2,11):
        num = 10 ** p
        i  = counter * (10**p)
        for pic in range(num):
            plt.scatter([curr_pt[0]], [curr_pt[1]], c='darkorange', s=6)
            curr_pt = gen_next(curr_pt)
        plt.scatter(origin_pts[...,0], origin_pts[...,1], color='steelblue', s=22)
        plt.title('i = ' + str(i), fontsize=50)
        plt.savefig('frames/'+str(p)+str((counter-1))+'.png')
plt.show()
```
