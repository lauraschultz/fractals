```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
current = [0,0]
```

```python
def get_new(ar):
    prob = np.random.random_sample()
    if (prob <0.01):
        ar[0] = 0
        ar[1] = 0.16*ar[1]
    elif (prob<0.86):
        ar[0] = (0.85*ar[0]) + (0.04*ar[1])
        ar[1] = (-0.04*ar[0]) + (0.85*ar[1]) + 1.6
    elif (prob < .93):
        ar[0] = (0.2*ar[0]) - (0.26*ar[1])
        ar[1] = (0.23*ar[0]) + (0.22*ar[1]) + 1.6
    else:
        ar[0] = (-.15*ar[0]) + (0.28*ar[1])
        ar[1] = (0.26*ar[0]) + (0.24*ar[1]) + .44
    return ar
```

```python
plt.clf()
plt.figure(figsize=(15,15))
plt.xlim(-3.1,3.9)
plt.ylim(-.1, 10.398)
plt.xticks([])
plt.yticks([])
#base case: when i=1
current = get_new(current)
plt.scatter([current[0]], [current[1]], c='forestgreen', s=1)
plt.title('i = 1', fontsize=50)
plt.savefig('frames_fern/00.png')
for p in range(0,5):
    for counter in range(2,11):
        num = 10 ** p
        i  = counter * (10**p)
        print("num is " + str(num) + ", i is " +str(i))
        for pic in range(num): 
            current = get_new(current)
            plt.scatter([current[0]], [current[1]], c='forestgreen', s=1)
        plt.title('i = ' + str(i), fontsize=50)
        plt.savefig('frames_fern/'+str(p)+str((counter-1))+'.png')
plt.show()
```

```python
#make animated visualizations 
import glob
import moviepy.editor as mpy

gif_name = 'fern.gif'
fps = 1
file_list = glob.glob('frames_fern/*.png') # Get all the pngs in the current directory
list.sort(file_list, key=lambda x: x.split('.')[0]) # Sort the images by #, this may need to be tweaked for your use case
clip = mpy.ImageSequenceClip(file_list, fps=3)
clip.write_gif(gif_name, fps=3)
    
#source: http://superfluoussextant.com/making-gifs-with-python.html
```
